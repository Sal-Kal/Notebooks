```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import *
from sklearn.model_selection import *
from sklearn.linear_model import LinearRegression
from sklearn.metrics import *
```

```python
data = pd.read_csv('./Advertising.csv')
data.head()
```

```python
data = data.drop(columns = 'Unnamed: 0', axis = 1)
data.head()
```

```python
xVar = data.drop(columns = 'sales', axis = 1)
yVar = data['sales'].values.reshape(-1,1)
%matplotlib inline
sns.pairplot(data, x_vars = xVar.columns, y_vars = 'sales', height = 7, aspect = 0.7, kind='reg')
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(xVar, yVar, test_size=0.3)
reg = LinearRegression()
reg.fit(xTrain, yTrain)
```

```python
print('The Equation of the line is: y = TV *',reg.coef_[0][0],' + Radio *',reg.coef_[0][1],' + Newspaper *',reg.coef_[0][2],'+',reg.intercept_[0])
```

```python
predictions = reg.predict(xTest)
mse = mean_squared_error(yTest, predictions)
mae = mean_absolute_error(yTest, predictions)
rmse = np.sqrt(mse)
print('The mean squared error is : ',mse)
print('The mean absolute error is : ',mae)
print('The root mean squared error is : ',rmse)
```
